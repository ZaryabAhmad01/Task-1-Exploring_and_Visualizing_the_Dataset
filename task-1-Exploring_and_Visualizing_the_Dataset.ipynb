```python
plt.figure()
plt.scatter(df["sepal_length"], df["sepal_width"])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter by species
setosa = df[df["species"] == "setosa"]
versicolor = df[df["species"] == "versicolor"]
virginica = df[df["species"] == "virginica"]

plt.figure()

plt.scatter(setosa["sepal_length"], setosa["sepal_width"], label="Setosa")
plt.scatter(versicolor["sepal_length"], versicolor["sepal_width"], label="Versicolor")
plt.scatter(virginica["sepal_length"], virginica["sepal_width"], label="Virginica")

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width by Species")
plt.legend()

plt.show()
```

```python
#  same task by using seaborn library
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df)
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()
```

```python
sns.scatterplot(x="petal_length", y="petal_width", hue="species", data=df)
plt.title("Petal Length vs Petal Width by Species")
plt.show()
```

```python
plt.figure()
plt.hist(df["sepal_length"], bins=20,color="purple",edgecolor="black")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
# Quick overview of all relationships
sns.pairplot(df, hue="species")
plt.show()
```

# Detect Outlier

```python
sns.distplot(df["petal_length"])
plt.show()
```

```python
# single colunms outliers checking 
sns.boxenplot(df["sepal_length"])
plt.show()
```

```python
# all features boxplot

plt.figure(figsize=(8,5))
sns.boxplot(data=df[["sepal_length","sepal_width","petal_length","petal_width"]])

plt.title("Boxplot of Iris Features")

#  if we want to save any chart by png format we can use this code 
plt.savefig("line_plot.png",dpi=300,bbox_inches="tight")
plt.show()
```
